```python
import matplotlib.pyplot as plt

# Esto creará una cuadrícula con todas tus 61 columnas
df.hist(figsize=(20, 15), bins=20, layout=(8, 8), color='skyblue', edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
# Filtramos solo las columnas que contienen la palabra "Mean"
mean_cols = [col for col in df.columns if 'Mean' in col]

df[mean_cols].hist(figsize=(12, 8), bins=15)
plt.suptitle("Distribución de Promedios de Sensores")
plt.show()
```

```python
# Filtramos solo las columnas que contienen la palabra "Mean"
acc_cols = [col for col in df.columns if 'Acc' in col]

df[acc_cols].hist(figsize=(12, 8), bins=15)
plt.subplots_adjust(hspace=2, wspace=0.3)
plt.suptitle("Distribucion de las aceleraciones en los distintos ejes")
plt.show()
```

```python
# 1. Separamos los datos por cada Target
clase1 = df[df['Target'] == 1]['AccMaxX']
clase2 = df[df['Target'] == 2]['AccMaxX']
clase3 = df[df['Target'] == 3]['AccMaxX']
clase4 = df[df['Target'] == 4]['AccMaxX']

fig, ax = plt.subplots(figsize=(8, 5))

# 2. Pasamos la lista de las 4 clases
ax.boxplot([clase1, clase2, clase3, clase4])

# 3. Ponemos las etiquetas de los comportamientos
ax.set_xticklabels(['Acelerón', 'Giro D.', 'Giro I.', 'Frenado'])
ax.set_title('Aceleración Máxima por Tipo de Maniobra')
plt.show()
```
